```python
from pyspark.sql.functions import col,lit
import dlt
```

```python
@dlt.table(name="02_silver.realestate_casablanca",comment="data enriched with metadata")
@dlt.expect_or_drop("valid price","new_price IS NOT NULL AND new_price > 0")
@dlt.expect_or_drop("valid city","city IS NOT NULL")
@dlt.expect_or_drop("valid surface","surface > 0 ")


def silver_RE_casa():
  return(
    dlt.readStream("01_bronze.Real_estate_casablanca")
       .withColumn("new_price",col("new_price").cast("double"))
       .withColumn("chambres",col("chambres").cast("int"))
       .withColumn("salles_de_bains",col("salles_de_bains").cast("int"))
       .withColumn("surface",col("surface").cast("int"))
       .withColumn("floor",col("floor").cast("int"))
       .withColumn("agency_id",lit("1"))
      )
```
