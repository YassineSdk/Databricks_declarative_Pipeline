```python
import dlt
@dlt.table(name="03_gold.Unified_RealEstate_data",comment="this table combines all the reale state data for the tree cities rabat ,casablanca , tanger ")
def combine_data():
    return (
        dlt.readStream("02_silver.realestate_rabat")
            .unionByName(dlt.readStream("02_silver.realestate_tanger"))
            .unionByName(dlt.readStream("02_silver.realestate_casablanca"))
    )
```
