```python
import pandas as pd 
from sdv.single_table import GaussianCopulaSynthesizer
from sdv.metadata import SingleTableMetadata
import pickle
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/Volumes/real_estate_management/sample_data/data_source/houses_data_eng.csv")
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df.rename({df.columns[2]:df.columns[2].replace(' ','_')},inplace=True)
```

```python
def train_model():
    save_path = Path("real_estate_management.sample_data.synthetic_data_generator")

    #defining meta data 
    metadata = SingleTableMetadata()
    metadata.detect_from_dataframe(df)

    # training the synthesize
    synthesizer = GaussianCopulaSynthesizer(metadata)
    synthesizer.fit(df)

    synthesizer.save(save_path)
    return synthesizer
train_model()
```

```python
import random
def generat_data():
    nb_rows = random.randint(1,30)

    path = Path("real_estate_management.sample_data.synthetic_data_generator")
    with open(path,'rb') as f:
        model = pickle.load(f)
    data = model.sample(nb_rows)
    return data ,nb_rows
data,rb_row  = generat_data()
print(data)
```
